```python
def find_guard(m):
    options = ['v','>','<','^']
    for i,line in enumerate(m):
        for option in options:
            if option in line:
                return i, line.index(option), option
            
def check_in_map(m,i,j):
    if i < 0 or j < 0:
        return False
    elif len(m) <= i:
        return False
    elif len(m[0]) <= j:
        return False
    else:
        return True
    
            
def check_free(m,i,j):
     return m[i][j] != '#'
    
def change_guard_direction(guard):
    directions = ['^','>','v','<']
    return directions[(directions.index(guard) + 1) % len(directions)]
```

```python
def up(m,i,j,guard):
    if not(check_in_map(m,i-1,j)):
        return -1
    if check_free(m,i-1,j):
        m[i][j] = 'X'
        m[i-1][j] = guard
        return True 
    else:
        return False

        
def right(m,i,j,guard):
    if not(check_in_map(m,i,j+1)):
        return -1
    if check_free(m,i,j+1):
        m[i][j+1] = guard
        m[i][j] = 'X'
        return True
    else:
        return False
        
def down(m,i,j,guard):
    if not(check_in_map(m,i+1,j)):
        return -1
    if check_free(m,i+1,j):
        m[i][j] = 'X'
        m[i+1][j] = guard
        return True
    else:
        return False

        
def left(m,i,j,guard):
    if not(check_in_map(m,i,j-1)):
        return -1
    if check_free(m,i,j-1):
        m[i][j] = 'X'
        m[i][j-1] = guard
        return True
    else:
        return False 

           
```

```python
directions = {
     '^' : up
    ,'>' : right
    ,'v' : down
    ,'<' : left 
}
first_position = find_guard(content)
```

# Part 1

```python
while True:
    guard = find_guard(content)
    i = guard[0]
    j = guard[1]
    move = directions[guard[2]]
    result = move(content, i, j,guard[2])
    if result == -1:
        break
    elif result is False:
        content[i][j] = change_guard_direction(guard[2])
content[i][j] = 'X'
```

```python
def solve_puzzle(content):
    register_move = []
    while True:
        guard = find_guard(content)
        if guard in register_move:
            return False
        register_move.append(guard)
        i = guard[0]
        j = guard[1]
        move = directions[guard[2]]
        result = move(content, i, j,guard[2])
        if result == -1:
            content[i][j] = 'X'
            return True
        elif result is False:
            content[i][j] = change_guard_direction(guard[2])
    
            
def get_all_x(content):
    x_position = []
    for i,valuei in enumerate(content):
        for j, valeuj in enumerate(valuei):
            if valeuj == 'X':
                x_position.append((i,j))
    return x_position

def put_obstacle(content, i,j):
    content[i][j] = '#'
def show_puzzle(content):
    for i in content:
        print("".join(i))
```
